```sql
-- Create a new table called '[Novedad]' in schema '[dbo]'
-- Drop the table if it already exists
IF OBJECT_ID('[dbo].[Novedad]', 'U') IS NOT NULL
    DROP TABLE [dbo].[Novedad]
GO

-- Create the table in the specified schema
CREATE TABLE [dbo].[Novedad] (
    Id_Novedad INT IDENTITY(1, 1) PRIMARY KEY
    , tipo_novedad VARCHAR(50)
    , motivo_novedad TEXT
    , fecha_registro VARCHAR(30)
    )
GO
```

```sql
-- Delete rows from table '[Rutas]' in schema '[dbo]'
DELETE
FROM [dbo].[Rutas]
WHERE nombre_ruta IN (
        SELECT nombre_ruta
        FROM Rutas
        GROUP BY nombre_ruta
        HAVING COUNT(*) > 1
        )
```
